```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def poly(x, theta):
    x_mat = np.array([x, ] * len(theta)).T
    x_pow = np.power(x_mat, range(0, len(theta)))
    return np.sum(theta * x_pow, axis=1)
```

```python
def get_error(y_p, h):
    return np.sum(np.square(y_p - h))
```

```python
def change():
    return 0.01 * (np.random.rand() - 0.5)
```

```python
x_points = np.array([1, 2, 3, 4])
y_points = np.array([1, 2, 1, 4])
```

```python
plt.plot(x_points, y_points, "ro")
```

```python
def train(theta, iters):
    
    best_error = np.inf
    theta_ = np.copy(theta)
    
    for i in range(iters):
        
        for t in range(0, theta.shape[0]):
            print("t", t)
            print("theta_[t]", theta_[t])
            print("theta[t]", theta[t])
            print("theta[t] + change()", theta[t] + change())
            theta_[t] = theta[t] + change()
            print("theta_[t]", theta_[t])
        
        print(theta_)
        h = poly(x_points, theta_)
        print(h)
        err = get_error(y_points, h)

        if(err < best_error):
            best_error = err;
            theta = theta_.copy()
            
        if((i % 500) == 0):
            print("iter: ", i, "err: ", err)
            
    print("h: ", poly(x_points, theta))
    print("y: ", y_points)
    print("error: ", get_error(y_points, h))
    
    return theta

    
```

```python
theta = np.array([0, 0, 0, 0])
theta = train(theta, 10000)
```
